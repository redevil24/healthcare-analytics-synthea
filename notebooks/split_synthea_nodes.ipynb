# 3. Th·ªëng k√™ danh s√°ch city trong `organizations.csv`

B·ªã v·∫•n ƒë·ªÅ hoa th∆∞·ªùng, BOSTON , boston, n√™n t√≠nh kh√°c.

```python
import pandas as pd

# üè• 1. L·∫•y b·∫£ng organizations g·ªëc
org = dfs["organizations"]  # Gi·ªØ nguy√™n d·ªØ li·ªáu g·ªëc

# üßπ 2. T·∫°o b·∫£n sao t·∫°m ƒë·ªÉ chia node
org_tmp = org.copy()

# üß≠ 3. T·∫°o c·ªôt CITY_CLEAN ƒë·ªÉ nh√≥m v√† chia node (ch·ªâ d√πng t·∫°m)
org_tmp["CITY_CLEAN"] = org_tmp["CITY"].astype(str).str.strip().str.lower()

# üßÆ 4. Th·ªëng k√™ s·ªë l∆∞·ª£ng t·ªï ch·ª©c theo CITY_CLEAN
city_count = (
    org_tmp.groupby("CITY_CLEAN")["Id"]
           .count()
           .reset_index()
           .rename(columns={"Id": "NUM_ORG"})
           .sort_values(by="NUM_ORG", ascending=False)
)

# üßæ 5. In k·∫øt qu·∫£
print("üèôÔ∏è Th·ªëng k√™ s·ªë l∆∞·ª£ng t·ªï ch·ª©c (organization) theo CITY_CLEAN ‚Äî d√πng ƒë·ªÉ chia node:")
display(city_count.head(20))

print(f"\nüìä T·ªïng s·ªë city (CITY_CLEAN): {city_count.shape[0]}")
print(f"üì¶ T·ªïng s·ªë t·ªï ch·ª©c (organization): {len(org_tmp)}")

# ‚úÖ Ghi ch√∫:
# - org: d·ªØ li·ªáu g·ªëc, kh√¥ng ƒë·ª•ng t·ªõi.
# - org_tmp: b·∫£n t·∫°m c√≥ th√™m CITY_CLEAN ƒë·ªÉ chia node.
# - Khi ghi file node, export theo c·ªôt g·ªëc t·ª´ org (kh√¥ng c√≥ CITY_CLEAN).
```

```python
city_count
```

```python
import math
import pandas as pd

# S·∫Øp city theo s·ªë l∆∞·ª£ng org gi·∫£m d·∫ßn
city_sorted = city_count.sort_values("NUM_ORG", ascending=False).reset_index(drop=True)

# T√≠nh s·ªë org m·ª•c ti√™u m·ªói node
total_org = city_sorted["NUM_ORG"].sum()
target = math.ceil(total_org / 8)

# Chia city th√†nh 8 nh√≥m ƒë∆°n gi·∫£n
node, node_sum, groups = 1, 0, {}
for _, row in city_sorted.iterrows():
    city, n = row["CITY_CLEAN"], row["NUM_ORG"]
    if node_sum + n > target and node < 8:
        node += 1
        node_sum = 0
    groups[city] = node
    node_sum += n

# T·ªïng k·∫øt k·∫øt qu·∫£
df_group = pd.DataFrame(list(groups.items()), columns=["CITY_CLEAN", "NODE"])
summary = df_group.groupby("NODE").size().reset_index(name="NUM_CITY")
summary["TOTAL_ORG"] = df_group.groupby("NODE")["CITY_CLEAN"].apply(
    lambda x: city_count[city_count["CITY_CLEAN"].isin(x)]["NUM_ORG"].sum()
).values

print(f"üéØ Target ~{target} org per node")
display(summary)

print("\nüó∫Ô∏è  Mapping m·∫´u (10 d√≤ng ƒë·∫ßu):")
display(df_group.head(10))
```

```python
# Gom city theo t·ª´ng node
node_summary = (
    city_count.merge(df_group, on="CITY_CLEAN")
              .sort_values(["NODE","NUM_ORG"], ascending=[True,False])
)

# G·ªôp danh s√°ch city (t√™n v√† s·ªë b·ªánh vi·ªán) trong m·ªói node
summary_table = (
    node_summary.groupby("NODE")
    .apply(lambda x: ", ".join([f"{c}({n})" for c,n in zip(x["CITY_CLEAN"], x["NUM_ORG"])]))
    .reset_index()
    .rename(columns={0: "CITY_LIST"})
)

# Th√™m t·ªïng s·ªë city & t·ªï ch·ª©c
summary_table["NUM_CITY"] = node_summary.groupby("NODE")["CITY_CLEAN"].count().values
summary_table["TOTAL_ORG"] = node_summary.groupby("NODE")["NUM_ORG"].sum().values

# S·∫Øp l·∫°i c·ªôt cho g·ªçn
summary_table = summary_table[["NODE", "NUM_CITY", "TOTAL_ORG", "CITY_LIST"]]

print("üìã Danh s√°ch city trong t·ª´ng node:")
display(summary_table)
```

# 4. Chia CSV th√†nh 8 node

```python
import os
import pandas as pd

# üîπ 1. Chu·∫©n b·ªã th∆∞ m·ª•c ƒë·∫ßu ra
OUT_DIR = "/content/nodes_csv"
os.makedirs(OUT_DIR, exist_ok=True)

# üîπ 2. D·ªØ li·ªáu g·ªëc
org = dfs["organizations"]
enc = dfs["encounters"]
cond = dfs["conditions"]
proc = dfs["procedures"]
med  = dfs["medications"]
pat  = dfs["patients"]

# üîπ 3. G·∫Øn NODE cho organizations theo CITY_CLEAN
org_tmp = org.copy()
org_tmp["CITY_CLEAN"] = org_tmp["CITY"].astype(str).str.strip().str.lower()
org_tmp = org_tmp.merge(df_group, on="CITY_CLEAN", how="left")

# Ki·ªÉm tra kh√¥ng c√≤n NaN NODE
assert org_tmp["NODE"].notna().all(), "‚ùå C√≥ t·ªï ch·ª©c ch∆∞a g√°n NODE!"

# üîπ 4. H√†m chia node
def export_node(node_id):
    node_path = f"{OUT_DIR}/node_{node_id}"
    os.makedirs(node_path, exist_ok=True)

    # a) organization c·ªßa node
    org_n = org_tmp[org_tmp["NODE"] == node_id]
    org_ids = set(org_n["Id"])

    # b) encounter thu·ªôc org ƒë√≥
    enc_n = enc[enc["ORGANIZATION"].isin(org_ids)]
    enc_ids = set(enc_n["Id"])

    # c) conditions / procedures / medications theo encounter
    cond_n = cond[cond["ENCOUNTER"].isin(enc_ids)]
    proc_n = proc[proc["ENCOUNTER"].isin(enc_ids)]
    med_n  = med[med["ENCOUNTER"].isin(enc_ids)]

    # d) patients trong encounter c·ªßa node
    pat_ids = set(enc_n["PATIENT"])
    pat_n = pat[pat["Id"].isin(pat_ids)]

    # e) Xu·∫•t file
    org_n.to_csv(f"{node_path}/organizations.csv", index=False)
    enc_n.to_csv(f"{node_path}/encounters.csv", index=False)
    cond_n.to_csv(f"{node_path}/conditions.csv", index=False)
    proc_n.to_csv(f"{node_path}/procedures.csv", index=False)
    med_n.to_csv(f"{node_path}/medications.csv", index=False)
    pat_n.to_csv(f"{node_path}/patients.csv", index=False)

    # f) Log k·∫øt qu·∫£
    print(f"\nüì¶ Node {node_id} exported:")
    print(f" - Organizations: {len(org_n):>6}")
    print(f" - Encounters:    {len(enc_n):>6}")
    print(f" - Conditions:    {len(cond_n):>6}")
    print(f" - Procedures:    {len(proc_n):>6}")
    print(f" - Medications:   {len(med_n):>6}")
    print(f" - Patients:      {len(pat_n):>6}")
    return {
        "node": node_id,
        "org": len(org_n),
        "enc": len(enc_n),
        "cond": len(cond_n),
        "proc": len(proc_n),
        "med": len(med_n),
        "pat": len(pat_n),
    }

# üîπ 5. Ch·∫°y cho t·∫•t c·∫£ 8 node
summary_nodes = []
for i in range(1, 9):
    summary_nodes.append(export_node(i))

# üîπ 6. T·ªïng k·∫øt (phi√™n b·∫£n s·ª≠a l·ªói)
summary_df = pd.DataFrame(summary_nodes)

# T√≠nh t·ªïng cho c√°c c·ªôt s·ªë (b·ªè c·ªôt 'node')
totals = summary_df.drop(columns=["node"]).sum(numeric_only=True)

# T·∫°o DataFrame 1 d√≤ng cho TOTAL
total_row = pd.DataFrame([["T·ªïng"] + totals.tolist()],
                         columns=["node"] + list(totals.index))

summary_df = pd.concat([summary_df, total_row], ignore_index=True)

print("\nüìä T·ªïng k·∫øt s·ªë l∆∞·ª£ng b·∫£n ghi t·ª´ng node:")
display(summary_df)
```

# 7. Validation sau chia (check to√†n v·∫πn & tr√πng l·∫∑p)

```python
import os
import pandas as pd

BASE_OUT = "/content/nodes_csv"
N_NODES = 8

def load_all(filename):
    dfs = []
    for i in range(1, N_NODES + 1):
        path = f"{BASE_OUT}/node_{i}/{filename}.csv"
        df = pd.read_csv(path)
        df["NODE"] = i
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

# ----------------------------------------------------------------------
# üß© 1. Cross-Node Validation
# ----------------------------------------------------------------------
print("üîé I. KI·ªÇM TRA LI√äN NODE\n")

# Load c√°c b·∫£ng c·∫ßn
org_all = load_all("organizations")
enc_all = load_all("encounters")
cond_all = load_all("conditions")
proc_all = load_all("procedures")
med_all = load_all("medications")
pat_all = load_all("patients")

# 1Ô∏è‚É£ Kh√¥ng tr√πng org.id
dupl_org = org_all.duplicated(subset="Id", keep=False).sum()
# 2Ô∏è‚É£ Kh√¥ng tr√πng enc.id
dupl_enc = enc_all.duplicated(subset="Id", keep=False).sum()

# 3Ô∏è‚É£ Kh√¥ng tr√πng condition/procedure/medication (d·ª±a tr√™n ENCOUNTER + CODE)
dupl_cond = cond_all.duplicated(subset=["ENCOUNTER","CODE"], keep=False).sum()
dupl_proc = proc_all.duplicated(subset=["ENCOUNTER","CODE"], keep=False).sum()
dupl_med  = med_all.duplicated(subset=["ENCOUNTER","CODE"], keep=False).sum()

# 4Ô∏è‚É£ patients c√≥ th·ªÉ tr√πng => ch·ªâ log s·ªë tr√πng
dupl_pat = pat_all.duplicated(subset="Id", keep=False).sum()

# 5Ô∏è‚É£ T·ªïng h·ª£p 8 node = t·ªïng g·ªëc
print(f"üì¶ Organizations t·ªïng: {len(org_all):>7}")
print(f"üì¶ Encounters t·ªïng:    {len(enc_all):>7}")
print(f"üì¶ Conditions t·ªïng:    {len(cond_all):>7}")
print(f"üì¶ Procedures t·ªïng:    {len(proc_all):>7}")
print(f"üì¶ Medications t·ªïng:   {len(med_all):>7}")
print(f"üì¶ Patients t·ªïng:      {len(pat_all):>7}\n")

print(f"‚ùå Tr√πng organization.id gi·ªØa node: {dupl_org}")
print(f"‚ùå Tr√πng encounter.id gi·ªØa node:    {dupl_enc}")
print(f"‚ùå Tr√πng condition (encounter+code): {dupl_cond}")
print(f"‚ùå Tr√πng procedure (encounter+code): {dupl_proc}")
print(f"‚ùå Tr√πng medication (encounter+code): {dupl_med}")
print(f"‚ÑπÔ∏è  patients tr√πng gi·ªØa node:       {dupl_pat} (ch·∫•p nh·∫≠n ƒë∆∞·ª£c)\n")

# 6Ô∏è‚É£ Ki·ªÉm tra patient ch∆∞a t·ª´ng c√≥ encounter
pat_enc_all = set(enc_all["PATIENT"])
pat_all_set = set(pat_all["Id"])
no_enc_pat = pat_all_set - pat_enc_all
print(f"üßç‚Äç‚ôÇÔ∏è B·ªánh nh√¢n ch∆∞a c√≥ encounter: {len(no_enc_pat)}")

# ----------------------------------------------------------------------
# üß± 2. Within-Node Validation
# ----------------------------------------------------------------------
print("\nüîé II. KI·ªÇM TRA N·ªòI NODE\n")

def validate_node(node_id):
    base = f"{BASE_OUT}/node_{node_id}"
    org = pd.read_csv(f"{base}/organizations.csv")
    enc = pd.read_csv(f"{base}/encounters.csv")
    cond = pd.read_csv(f"{base}/conditions.csv")
    proc = pd.read_csv(f"{base}/procedures.csv")
    med  = pd.read_csv(f"{base}/medications.csv")
    pat  = pd.read_csv(f"{base}/patients.csv")

    # FK checks
    bad_org  = enc.loc[~enc["ORGANIZATION"].isin(org["Id"])]
    bad_pat  = enc.loc[~enc["PATIENT"].isin(pat["Id"])]
    bad_cond = cond.loc[~cond["ENCOUNTER"].isin(enc["Id"])]
    bad_proc = proc.loc[~proc["ENCOUNTER"].isin(enc["Id"])]
    bad_med  = med.loc[~med["ENCOUNTER"].isin(enc["Id"])]

    # Patient consistency (conditions/procedures/medications ‚Üí patients)
    bad_cond_pat = cond.loc[~cond["PATIENT"].isin(pat["Id"])]
    bad_proc_pat = proc.loc[~proc["PATIENT"].isin(pat["Id"])]
    bad_med_pat  = med.loc[~med["PATIENT"].isin(pat["Id"])]

    # Organization ch∆∞a c√≥ encounter
    org_has_enc = set(enc["ORGANIZATION"].unique())
    org_all = set(org["Id"].unique())
    org_without_enc = org_all - org_has_enc

    return {
        "node": node_id,
        "bad_org": len(bad_org),
        "bad_pat": len(bad_pat),
        "bad_cond": len(bad_cond),
        "bad_proc": len(bad_proc),
        "bad_med": len(bad_med),
        "bad_cond_pat": len(bad_cond_pat),
        "bad_proc_pat": len(bad_proc_pat),
        "bad_med_pat": len(bad_med_pat),
        "org_without_enc": len(org_without_enc),
    }

fk_results = pd.DataFrame([validate_node(i) for i in range(1, N_NODES+1)])
display(fk_results)

# ----------------------------------------------------------------------
# üßÆ T·ªïng k·∫øt Pass/Fail
# ----------------------------------------------------------------------
summary_cols = [c for c in fk_results.columns if c != "node"]
fail = fk_results[summary_cols].sum().sum()

if fail == 0 and dupl_org==0 and dupl_enc==0:
    print("\n‚úÖ T·∫§T C·∫¢ KI·ªÇM TRA PASS ‚Äî D·ªÆ LI·ªÜU CHIA H·ª¢P L·ªÜ 100%.")
else:
    print("\n‚ö†Ô∏è C√ì L·ªñI ‚Äî vui l√≤ng xem b·∫£ng chi ti·∫øt tr√™n.")
```

# 6. Convert to√†n b·ªô node CSV ‚Üí Parquet (theo schema g·ªëc)

```python
import os
import pandas as pd

SRC_DIR = "/content/nodes_csv"
DST_DIR = "/content/nodes_parquet"
os.makedirs(DST_DIR, exist_ok=True)

# ======================
# SCHEMA CHU·∫®N (Datasource_Dictionary.docx)
# ======================

schema_patients = {
    "Id": "string", "BIRTHDATE": "string", "DEATHDATE": "string",
    "SSN": "string", "DRIVERS": "string", "PASSPORT": "string",
    "PREFIX": "string", "FIRST": "string", "LAST": "string", "SUFFIX": "string",
    "MAIDEN": "string", "MARITAL": "string", "RACE": "string", "ETHNICITY": "string",
    "GENDER": "string", "BIRTHPLACE": "string", "ADDRESS": "string", "CITY": "string",
    "STATE": "string", "COUNTY": "string", "ZIP": "string",
    "LAT": "float64", "LON": "float64",
    "HEALTHCARE_EXPENSES": "float64", "HEALTHCARE_COVERAGE": "float64"
}

schema_encounters = {
    "Id": "string", "START": "string", "STOP": "string",
    "PATIENT": "string", "ORGANIZATION": "string", "PROVIDER": "string",
    "PAYER": "string", "ENCOUNTERCLASS": "string", "CODE": "string",
    "DESCRIPTION": "string", "BASE_ENCOUNTER_COST": "float64",
    "TOTAL_CLAIM_COST": "float64", "PAYER_COVERAGE": "float64",
    "REASONCODE": "string", "REASONDESCRIPTION": "string"
}

schema_conditions = {
    "START": "string", "STOP": "string", "PATIENT": "string",
    "ENCOUNTER": "string", "CODE": "string", "DESCRIPTION": "string"
}

schema_procedures = {
    "DATE": "string", "PATIENT": "string", "ENCOUNTER": "string",
    "CODE": "string", "DESCRIPTION": "string", "BASE_COST": "float64",
    "REASONCODE": "string", "REASONDESCRIPTION": "string"
}

schema_medications = {
    "START": "string", "STOP": "string", "PATIENT": "string",
    "PAYER": "string", "ENCOUNTER": "string", "CODE": "string",
    "DESCRIPTION": "string", "BASE_COST": "float64", "PAYER_COVERAGE": "float64",
    "DISPENSES": "Int64", "TOTALCOST": "float64",
    "REASONCODE": "string", "REASONDESCRIPTION": "string"
}

schema_organizations = {
    "Id": "string", "NAME": "string", "ADDRESS": "string", "CITY": "string",
    "STATE": "string", "ZIP": "string", "LAT": "float64", "LON": "float64",
    "PHONE": "string", "REVENUE": "float64", "UTILIZATION": "float64"
}

schemas = {
    "patients": schema_patients,
    "encounters": schema_encounters,
    "conditions": schema_conditions,
    "procedures": schema_procedures,
    "medications": schema_medications,
    "organizations": schema_organizations
}

# ======================
# PH√ÇN LO·∫†I C·ªòT DATE vs DATETIME
# ======================

# date-only
cols_date = {
    "patients": ["BIRTHDATE", "DEATHDATE"],
    "conditions": ["START", "STOP"]
}

# datetime (c√≥ gi·ªù)
cols_datetime = {
    "encounters": ["START", "STOP"],
    "procedures": ["DATE"],
    "medications": ["START", "STOP"]
}

# ======================
# H√ÄM CHUY·ªÇN CSV ‚Üí PARQUET
# ======================

for node_id in range(1, 9):
    src_node = f"{SRC_DIR}/node_{node_id}"
    dst_node = f"{DST_DIR}/node_{node_id}"
    os.makedirs(dst_node, exist_ok=True)

    print(f"\nüì¶ Converting Node {node_id} ...")

    for name, schema in schemas.items():
        src_file = f"{src_node}/{name}.csv"
        dst_file = f"{dst_node}/{name}.parquet"

        # ƒê·ªçc CSV v·ªõi schema ƒë·ªãnh nghƒ©a
        df = pd.read_csv(src_file, dtype=schema, na_values=["", "NaN", None])

        # Convert c·ªôt date-only
        if name in cols_date:
            for col in cols_date[name]:
                if col in df.columns:
                    df[col] = pd.to_datetime(df[col], errors="coerce").dt.date

        # Convert c·ªôt datetime
        if name in cols_datetime:
            for col in cols_datetime[name]:
                if col in df.columns:
                    df[col] = pd.to_datetime(df[col], errors="coerce")

        # Ghi ra Parquet
        df.to_parquet(dst_file, index=False)
        print(f" - {name}.parquet ‚úÖ ({len(df)} rows)")

print("\nüéâ Ho√†n t·∫•t! T·∫•t c·∫£ 8 node ƒë√£ ƒë∆∞·ª£c convert sang Parquet typed ƒë√∫ng chu·∫©n trong /content/nodes_parquet/")
```

Ki·ªÉm tra schema th·ª±c t·∫ø trong Parquet c·ªßa c√°c node

```python
import pyarrow.parquet as pq
import os

PARQUET_DIR = "/content/nodes_parquet"

def check_schema(node_id):
    node_dir = f"{PARQUET_DIR}/node_{node_id}"
    print(f"\nüß© Schema of Node {node_id}")
    for file in sorted(os.listdir(node_dir)):
        if file.endswith(".parquet"):
            path = os.path.join(node_dir, file)
            schema = pq.read_table(path).schema
            print(f"\nüìò {file}")
            print(schema)

# ‚úÖ Ki·ªÉm tra to√†n b·ªô 8 node
for node in range(8, 9):
    check_schema(node)
```

T√°i c·∫•u tr√∫c th∆∞ m·ª•c Parquet theo chu·∫©n Iceberg

```python
import os, shutil

SRC_DIR = "/content/nodes_parquet"
DST_DIR = "/content/nodes_parquet_iceberg"
os.makedirs(DST_DIR, exist_ok=True)

print("üóÇÔ∏è ƒêang t√°i c·∫•u tr√∫c th∆∞ m·ª•c th√†nh d·∫°ng Iceberg...\n")

for node_id in range(1, 9):
    src_node = f"{SRC_DIR}/node_{node_id}"
    dst_node = f"{DST_DIR}/node_{node_id}"
    os.makedirs(dst_node, exist_ok=True)

    # Li·ªát k√™ c√°c file .parquet trong node
    parquet_files = [f for f in os.listdir(src_node) if f.endswith(".parquet")]
    print(f"üì¶ Node {node_id}: {len(parquet_files)} b·∫£ng")

    for f in parquet_files:
        table_name = f.replace(".parquet", "")
        src_path = os.path.join(src_node, f)
        dst_table_folder = os.path.join(dst_node, table_name)
        os.makedirs(dst_table_folder, exist_ok=True)

        # Copy file v√†o th∆∞ m·ª•c ri√™ng
        dst_path = os.path.join(dst_table_folder, f)
        shutil.copy2(src_path, dst_path)
        print(f"  ‚îú‚îÄ‚îÄ {table_name}/ ({f}) ‚úÖ")

print("\nüéâ Ho√†n t·∫•t! C·∫•u tr√∫c m·ªõi n·∫±m t·∫°i:", DST_DIR)
```
